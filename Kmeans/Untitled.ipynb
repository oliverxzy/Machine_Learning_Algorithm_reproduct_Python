```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Euclidean_distance(var1, var2):
    squared_dis = 0
    for i in range(len(var1)):
        squared_dis += (var1[i] - var2[i]) ** 2

    return np.sqrt(squared_dis)
```

```python
Euclidean_distance([1,1,3],[1,1,7])
```
