```python
import numpy as np
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
x, y, z = np.random.multivariate_normal([10,10,10],[[1,0,0],[0,1,0],[0,0,1]],500).T
x2, y2, z2 = np.random.multivariate_normal([5,5,5],[[1,0,0],[0,1,0],[0,0,1]],500).T
x3, y3, z3 = np.random.multivariate_normal([20,20,20],[[1,0,0],[0,1,0],[0,0,1]],500).T
```

```python
fig = px.scatter_3d(x=x,y=y,z=z)
fig.show()
```

```python
plt.scatter(z,y)
plt.show()
```

```python
frame = {'x':np.concatenate((x,x2,x3)).tolist(),
         'y':np.concatenate((y,y2,y3)).tolist(),
         'z':np.concatenate((z,z2,z3)).tolist()}
df = pd.DataFrame(frame, columns=['x','y','z'])
```

```python
df.as_matrix()
```

```python
np.mean(df.as_matrix())
```

```python
np.cov(df.as_matrix())
```

```python
df.to_csv('pca_data.csv',index=False)
```
